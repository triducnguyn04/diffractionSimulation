```python
import numpy as np
import matplotlib.pyplot as plt
s, I = np.loadtxt('bragg_pattern.txt', delimiter=',', skiprows=1, unpack=True)
plt.stem(s, I, linefmt='b-', basefmt='k-', markerfmt='none')

plt.xlabel('s (Å⁻¹)')
plt.ylabel('Intensity')
plt.title('Bragg Diffraction Pattern')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from diffraction_pattern.txt
s, I = np.loadtxt('diffraction_pattern.txt', delimiter=',', skiprows=1, unpack=True)

plt.figure(figsize=(12, 6))
plt.plot(s, I, 'b-', label='Debye Powder (Original)')
plt.xlabel('s (Å⁻¹)')
plt.ylabel('Intensity')
plt.title('Xray Diffraction Pattern (25 Angstrom VO₂ Powder)')
plt.grid(True)
plt.legend()
plt.show()
```
